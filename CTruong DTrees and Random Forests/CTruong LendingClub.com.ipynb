In this project I explore real world, publicly available data from LendingClub.com. LendingClub.com is a website that conects borrowers with investors. I use Decision Trees and Random Forest Machine Learning algorithms to create models that help predict profiles that have a high probability of paying investors back. 




Here are what the columns represent:

credit.policy: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.

purpose: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").

int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.

installment: The monthly installments owed by the borrower if the loan is funded.

log.annual.inc: The natural log of the self-reported annual income of the borrower.

dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).

fico: The FICO credit score of the borrower.
days.with.cr.line: The number of days the borrower has had a credit line.

revol.bal: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).

revol.util: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).

inq.last.6mths: The borrower's number of inquiries by creditors in the last 6 months.

delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.

pub.rec: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

The column we are trying to predict is the not.fully.paid column.

# Setting up Data for Random Forest Classification Model

```python
loans.info()
```

# Categorical Features

The purpose column of our dataset is indicated as categorical, where as all others are integers/floats. We need to transform the values through the use of dummy variables so that sklearn will be able to understand.

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

# Train Test Split

```python
from sklearn.model_selection  import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)

y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Training a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

# Predictions and Evaluation of Decision Tree

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

From these predictions and evaluations, we have a classification report reporting precisiona and recall of ~75% depending on what classes 0 vs 1 we are looking at.

# Training the Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train,y_train)
```

# Predictions and Evaluations for Random Forest Model

```python
predictions = rfc.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

As can be seen, the precision recall and f1 score for certain classes improved and overall on average/total also improved but the recall in class 1 for instance is ~.03 and it was ~.24 for the single decision tree. So which performed better? The single decision tree performed better on class 1 for recall than our random forest. Also for class 1 for the f1 score it was a score of ~.06 whereas the single decision tree had a score of ~.22. Therefore, it really depends on what the costs associated with the calls is. 

Overall with the average the random forest model performed better but by looking at specific areas it performed a lot worse. Thereofore to make full informed evaluations, more context and business consulting would be required.
