In this Project I perform a K Means Clustering Machine Learning algorithm

K Means clustering is an unsupervised learning algorithm and clusters data based on their similarities

I am not looking for an outcome but rather I am looking for grouping/labels

Artificial data used

Process:
Import KMeans

Instantiate kmeans with clusters that i want to explore or expect

Fit it to the data

report back to cluster centers and label

compare the label predictions from the algorithm to actual labels. (K means is unsupervised so in real world data labeled data would not be available.)

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

Creating artificial data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')   #visualizing clusters and data[:,0] grab all the row in the first column 
```

```python
from sklearn.cluster import KMeans
```

Algorithm starts by randomly assigning each observation to a cluster and then find a centroid for each cluster

then the algorithms iterates to first two steps reassigns data points to the centroid whos the closest 

calculates the new centroid for each cluster and repeats until cluster variation can no longer be reduced any more

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_ 
```

```python
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

1. Create sub plot of 1 row by two columns 
2. Share=True allows them to share the same axis
3. I use tuple unpacking to grab both the axis
4. The first set of axis is going to plot the original data and color it based on what the algorithm predicted the labels should look like
5. My other axis is going to plot the original data but color them by their actually correct labels

from the visualization, it can be seen that there is noise in the original plot; however, the algorithm could figure it out and assigned them all to the same centroid cluster.
